```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 데이터 준비
file_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
file_data = pd.read_csv(file_path)
file_data.head()
```

```
# 종속, 독립변수
inV = file_data[['온도']]
deV = file_data[['판매량']]
print(inV.shape, deV.shape)
```

```
# 모델 만들기
X = tf.keras.layers.Input(shape=[1])    #독립변수의 수 1개니까 shape=[1]
Y = tf.keras.layers.Dense(1)(X)          #종속변수 1개 니까
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')               #모델이 학습 할 방법
```

```
# 모델 학습 시키기
model.fit(inV,deV,epochs=10000,verbose=0) # 가져온 데이터 넣기, verbose=0은 학습상황 안보여줌
```

```
# 만번 학습 확인해보기 loss값
model.fit(inV,deV,epochs=10) # 가져온 데이터 넣기
```

```
# 모델 이용
model.predict(inV)
```

```
# 실제 데이터와 모델이 예측한 값 비교해보기
deV
```

```
model.predict([[15]]) 
```

```
# 과거 데이터 준비
iris_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris_data = pd.read_csv(iris_path)
iris_data.head()
```

```
#원핫인코딩/ 회귀가 아닌 분류, 분류를 위해 각 상태?를 숫자로
incoding_data = pd.get_dummies(iris_data)
incoding_data.head()
```

```
print(incoding_data.columns)
print(incoding_data.shape)
```

```
# 독립, 종속변수 
iris_inV = incoding_data[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
iris_deV = incoding_data[['품종_setosa', '품종_versicolor',
       '품종_virginica']]
print(iris_inV.shape, iris_deV.shape)
```

```
# 모델 학습시키기
model.fit(iris_inV, iris_deV, epochs=100)
```

```
##히든레이어, 아리리스 구분하기, 분류
import tensorflow as tf
import pandas as pd

# 과거 데이터 준비
iris_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris_data = pd.read_csv(iris_path)
iris_data.head()
```

```
#원핫인코딩/ 회귀가 아닌 분류, 분류를 위해 각 상태?를 숫자로
incoding_data = pd.get_dummies(iris_data)
incoding_data.head()

print(incoding_data.columns)
print(incoding_data.shape)
```

```
# 독립, 종속변수 
iris_inV = incoding_data[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
iris_deV = incoding_data[['품종_setosa', '품종_versicolor',
       '품종_virginica']]
print(iris_inV.shape, iris_deV.shape)
```

```
# 모델 구조 만들기 
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation='swish') (X)
H = tf.keras.layers.Dense(8, activation='swish') (H)
H = tf.keras.layers.Dense(8, activation='swish') (H)
Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy',metrics='accuracy')
```

```
# 모델 학습시키기
model.fit(iris_inV, iris_deV, epochs=100)
```

```
# 모델 이용하기
print(model.predict(iris_inV[0:5]))
print(iris_deV[0:5])
```

```
#성능 좋은 모델 이용한, 보스턴 집값 예측하기

import tensorflow as tf
import pandas as pd

boston_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv';
boston_data = pd.read_csv(boston_path)
print(boston_data.columns)
print(boston_data.head())



#독립, 종속변수 분리하기
boston_inV = boston_data[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
boston_deV = boston_data[['medv']]
print(boston_inV.shape, boston_deV.shape)
```

```
# 성능좋은 모델 만들기 Batch
X = tf.keras.layers.Input(shape=[13])

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H) # Dense랑 activation사이에 
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```
#학습시키기
model.fit(boston_inV, boston_deV, epochs=1000)
```

```
model.fit(boston_inV, boston_deV, epochs=10)
```
